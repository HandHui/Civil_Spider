```python
import time
```

```python
pre_ht = 'http://bjgy.chinacourt.gov.cn'
```

```python
def get_hrefs(page1):
    pat = r'<div id="list" class="font14">(.*?)<div class="paginationControl">'
    ## pat = '<div id="list"(.*?)<a href="/paper/detail/2013/11/id/1265994.shtml'
    case_list_raw = re.search(pat,page1,re.S)   ### .这个通配符匹配不了换行符
    case_list_raw = case_list_raw.groups()[0].strip()

    hrefs_pat = r'<a href="(.*?)"'
    hrefs = re.findall(hrefs_pat,case_list_raw)
#     print(len(hrefs))
    
    return hrefs

    #print("case_list",case_list.groups()[0])
```

```python
###  text为page的所有内容  text = browser.page_source
###  输出为文本的原先的格式

def parser(text):

    pat = '<li class="title">(.*?)</li>'
    title = re.search(pat,text)
    title = title.groups()[0]
    #     print("title",title.groups()[0])

    pat2 = '<div><p>(.*?)</p></div> '
    res = re.search(pat2,text)

    res = res.groups()[0].split('。',1)

    res[1] = res[1].replace('&nbsp;','') 

    para_pat = '<br>(.*?)<br>'
    para = re.finditer(para_pat,res[1])
    
    title1 = ''
    if len(title) >= 75:
        title1 = title[:75]
    else:
        title1 = title

    with open('case/BeiJing/'+title1+'.txt','w+',encoding = 'utf-8') as f:
        f.write(title+'\n')
        for p in para:
            p = p.groups()[0]
            p = p.strip('</p>').strip()
            if p == '':
                continue
#             print(p)
            f.write(p+'\n')
    f.close()
```

```python
###拿到第一页的列表
browser = webdriver.Firefox(log_path = "webdriver1.log")

for i in range(251,314):
    print('****************************')
    try:
        page_num = "http://bjgy.chinacourt.gov.cn/paper/more/paper_mid/MzA0gAMA/page/{}.shtml".format(i)
        print(i)
        time.sleep(0.3)

        browser.get(page_num)
        page1 = browser.page_source

        hrefs = get_hrefs(page1)
#         print("hrefs",hrefs)

        for href in hrefs:

            browser.get(pre_ht+href)
            time.sleep(0.3)
            ##rowser.get(t)
            text = browser.page_source
#             print('***************************')
            parser(text)
#             print('***************************')
    except AttributeError:
            continue

browser.close()
```
