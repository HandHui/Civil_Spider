```python
def parser(text2):
    title_pat = '<h2>(.*?)</h2>'
    title = re.search(title_pat,text2)

    title = title.groups()[0].strip()

    con_pat = '<p style="text-indent:30.0pt;"><span style="font-family:.*?;font-size:16.0pt;">(.*?)</span></p>'

    content = re.findall(con_pat,text2,re.S)
    
    title1 = ''
    if len(title) >= 75:
        title1 = title[:75]
    else:
        title1 = title
    
    with open('case/guangdong/guanggao/civil1/'+title1+'.txt','w+',encoding='utf-8') as f:
#         print("open a document")
        f.write(title+'\n')
        for con in content:
            f.write(con+'\n')
    f.close()
```

```python
browser = webdriver.Firefox(log_path = "webdriver.log")

for i in range(100,1001):
    list_page = 'http://www.gdcourts.gov.cn/index.php?m=openness&f=index&v=ws_list&ah=&fjm=J00&fyjc=%E5%B9%BF%E4%B8%9C%E9%AB%98%E9%99%A2&ay=&wslb=&ajlb=ms&wsmc=&page={}'.format(i)
    print(list_page)
    try:
        browser.get(list_page)
        time.sleep(0.3)
#     except:
#         print('***************************')
#         print(i)
#         print('****************************')
#         continue
        text = browser.page_source

        pat = '<td colspan="3" class="title">(.*?)</td>'
        res = re.findall(pat,text,re.S)

        for rg in res:
            hr = 'href="(.*?)"'
            a = re.search(hr,rg)

            aa = a.groups()

            browser.get(aa[0])
            time.sleep(0.5)
            text2 = browser.page_source
            parser(text2)
    except:
        print('********************')
        print(i)
        print('********************')
        i = i-1
#         break
#     break
#         print(browser.page_source)
    
# #     print(browser.page_source)
browser.close()

# pat = '<td colspan="3" class="title">(.*?)</td>'
# res = re.findall(pat,text,re.S)

# for rg in res:
#     print(rg)

# # res.groups()
```
